## What type of language is python?

[Python](python.org) is an interpreted high level dynamically typed garbage-collected language. This means, inlike java or C, you don't need to compile python code.  

* *Interpreted*: You interact with the language via a [REPL](https://en.wikipedia.org/wiki/Read%E2%80%93eval%E2%80%93print_loop).
* *High level*: Functions themselves are objects that can be passed around.
* *Dynamically typed*: variables don't have types, values do!
* *Garbage collected*: if you create an object, and you don't need it anymore, it is automatically erased.

## Comprehensions

Comprehensions is a useful idiom that makes a lot of things easy:

Say you have an array,

```python
a = [-1,0,2,-1,2,0,-2,3,4,0,0,2,3,4]
```

```python
{ x*x for x in a if x > 0 }
```

```python
xs = ['a', 'b', 'c']
ys = [1, 2, 3]
zs = {x:y for x,y in zip(xs,ys)}
```

## Functions

A function is an object that takes inputs and produces an output.



```python
def fibonacci(n):
    res = [1,1]
    for i in range(n-2):
        res.append(res[i] + res[i+1])
    return(res)
```

# A simple example with NumPY

In the first example we are going to write, we are going to approximate $\pi$ using random numbers: the probability that a uniformly randomly chosen point in the square determined by $(0,0)$ and $(1,1)$ that lies in the quarter of a  disk $x^2+y^2\leq 1$ is $\frac{\pi}{4}$.



```python
count = 0
N = 500000
for i in range(N):
    x = np.random.rand()
    y = np.random.rand()
    if(x**2 + y**2 <= 1):
        count = count + 1
count*4/N
```

```python
count = 0
N = 500000
points = np.random.rand(N,2)
for i in range(N):
    if( points[i,0]**2 + points[i,1]**2 <= 1):
        count = count + 1
count*4/N
```

```python
def circle(x): return(x[0]**2 + x[1]**2 <= 1)

N = 4000
points = np.random.rand(N,2)
count = ['a' for i in range(N) if circle(points[i,:])]
len(count)*4/N
```

```python
def area(N, fn):
    points = np.random.rand(N,2)
    count = len([1 for i in range(N) if fn(points[i,:])])
    return(count/N)
```

```python
4*area(100, circle)
```

```python
def line(x): return(x[1]<=x[0])

area(4000,line)
```

```python
area(5000, lambda x: x[1]<=x[0])
```

```python
area(5000, lambda x: x[1]<=x[0]**2)
```

```python
area(5000, lambda x: x[0]**2+x[1]**2<=1)*4
```

# Another example (exchange rates) with Pandas



```python
from urllib.request import urlopen
from xmltodict import parse
from pandas import DataFrame
```

```python
with urlopen("http://www.ecb.europa.eu/stats/eurofxref/eurofxref-daily.xml") as conn:
    raw = parse(conn.read().decode('utf8'))
raw
```

```python
data = raw['gesmes:Envelope']['Cube']['Cube']['Cube']
data
```

```python
rates = {x['@currency']: float(x['@rate']) for x in data}
rates
```

```python
rates['TRY']
```

```python
def getExchangeRates():
    with urlopen("http://www.ecb.europa.eu/stats/eurofxref/eurofxref-daily.xml") as conn:
        raw = parse(conn.read().decode('utf8'))['gesmes:Envelope']['Cube']['Cube']['Cube']
    rates = {x['@currency']: float(x['@rate']) for x in raw}
    return(rates)

rates = getExchangeRates()
```

```python
rates['USD']
```

```python
def QueryECB(Cfrom, Cto):
    with urlopen("https://sdw-wsrest.ecb.europa.eu/service/data/EXR/D." + Cfrom + "." + Cto + ".SP00.A") as conn:
        raw = parse(conn.read().decode('utf8'))
    data = raw['message:GenericData']['message:DataSet']['generic:Series']['generic:Obs']
    dates = [x['generic:ObsDimension']['@value'] for x in data]
    res = [float(x['generic:ObsValue']['@value']) for x in data]
            
    return DataFrame({f"{Cfrom}-{Cto}": res}, index=dates)

TRYEUR = QueryECB('TRY','EUR')
```

```python
df = DataFrame(TRYEUR)
df.plot(figsize=(20,6))
```

```python
USDEUR = QueryECB('USD','EUR')
df = DataFrame(USDEUR)
df.plot(figsize=(20,6))
```

```python
with urlopen("https://sdw-wsrest.ecb.europa.eu/service/data/LFSI/M.I8.S.UNEHRT.TOTAL0.15_74.T?format=genericdata") as conn:
    raw = parse(conn)
    
data = raw['message:GenericData']['message:DataSet']['generic:Series']['generic:Obs']
```

```python
rates = [float(x['generic:ObsValue']['@value']) for x in data]
dates = [x['generic:ObsDimension']['@value'] for x in data]

df = DataFrame({'Unemployment': rates}, index = dates)

df.plot(figsize=(20,6))
```

# Yet Another Example (Iris Dataset) with scikit-learn

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
y
```

```python
model = PCA(n_components=2)
res = model.fit_transform(X)

fig = plt.figure(figsize=(12,6))
plt.scatter(res[:,0],res[:,1],c=y)
plt.show()
```

```python
model = PCA(n_components=3)
res = model.fit_transform(X)

fig = plt.figure(figsize=(8,8))

ax = Axes3D(fig, elev=-120, azim=120, auto_add_to_figure=False)
ax.scatter(res[:,0],res[:,1],res[:,2],c=y,cmap=plt.cm.Set1)
fig.add_axes(ax)

plt.show()
```
