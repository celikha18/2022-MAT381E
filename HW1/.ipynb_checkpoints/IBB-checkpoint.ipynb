```python
from urllib.request import urlopen
import pandas as pd
from xmltodict import parse
from functools import reduce 
```

```python
with urlopen("https://data.ibb.gov.tr/en/datastore/odata3.0/604776d6-e99f-469c-bf25-25ccadc5e89b") as url:
    raw = url.read().decode('utf8')
```

```python
data = parse(raw)
```

```python
[x['content']['m:properties']['d:LATITUDE'] for x in data['feed']['entry']]
```

```python
tmp = [x['content']['m:properties'] for x in data['feed']['entry']]
res = []
i = 0
for x in tmp:
    try:
        name = x["d:STATION_NAME"]['#text']
        count = int(x["d:PASSENGER_COUNT"]['#text'])
        lat = float(x["d:LATITUDE"]["#text"])
        long = float(x["d:LONGITUDE"]["#text"])
        res.append({"name": name, "count": count, "lat": lat, "long": long})
    except:
        i = i+1
        
metro = pd.DataFrame(res)
metro
```
